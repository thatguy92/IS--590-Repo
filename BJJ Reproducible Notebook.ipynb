The code immediately below is used to import pandas which is the python package that I used to mask my intermediate data sets together.

```python
import pandas as pd
```

The following three lines of code allow pandas to read the datasets from my computer. These datasets are saved as csv hence the "read_csv" part of the code. The full path to the file must be provided in the brackets.

```python
SJJIF = pd.read_csv('https://raw.githubusercontent.com/thatguy92/IS--590-Repo/master/SJJIF%202018%20Worlds%20Data.csv')
```

```python
IBJJF = pd.read_csv('https://raw.githubusercontent.com/thatguy92/IS--590-Repo/master/IBJJF%202018%20Worlds%20Combined%20Data.csv')
FIVE = pd.read_csv('https://raw.githubusercontent.com/thatguy92/IS--590-Repo/master/2018%20Five%20Grappling%20Data%204.13.csv')
```

The following two lines of code allow me to print out the dimensions of the of the datasets as read by pandas. This ensures that the size remains the same after the reading.

```python
print(SJJIF.shape)
print(IBJJF.shape)
```

The following line of code combines the IBJJF dataset to the SJJIF dataset so that they can be analyzed together as one file.

```python
appended_data = SJJIF.append(IBJJF)
```

The following line of code measures the dimensions of the appended_data dataset created from the last line of code. It makes sure that it is the right size.

```python
appended_data.shape
```

The following code appends the FIVE Grappling dataset to the already created appended_data dataset so that all three datasets can be analyzed as one.

```python
appended_data = appended_data.append(FIVE)
```

The following code checks the dimensions of the new appended_data dataset to make sure that it accounts for all three datasets.

```python
appended_data.shape
```

The following line of code prints out the full appended_data dataset so that I can make sure that it counts for the data from all datasets.
